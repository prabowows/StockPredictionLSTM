```python
import yfinance as yf
import pandas as pd
from datetime import datetime, timedelta

def get_indonesian_stock_tickers():
    """
    Fungsi ini seharusnya mengambil daftar ticker saham Indonesia.
    Karena yfinance tidak menyediakan daftar ini secara langsung,
    Anda perlu mendapatkan daftar ini dari sumber lain.
    
    Sebagai contoh, saya akan menggunakan daftar kecil.
    Dalam aplikasi nyata, Anda bisa:
    1. Mengunduh dari situs IDX (seringkali dalam bentuk CSV/Excel).
    2. Menggunakan API pihak ketiga (jika ada yang menyediakan daftar ticker).
    3. Mengikis (scrape) dari situs web keuangan (hati-hati dengan legalitas dan terms of service).
    """
    
    # --- Contoh Daftar Ticker Saham Indonesia ---
    # Ini adalah daftar kecil dan mungkin tidak lengkap atau selalu up-to-date.
    # Anda harus menggantinya dengan daftar yang lebih komprehensif jika tersedia.
    return [
        "BBCA.JK", "BBRI.JK", "TLKM.JK", "ASII.JK", "BMRI.JK", 
        "GOTO.JK", "BBNI.JK", "AMZN", "MSFT", "GOOG" # Tambahan beberapa saham AS untuk demo multi-country
    ]

def get_stock_data(tickers, start_date, end_date, interval="1d"):
    """
    Mengunduh data historis untuk daftar ticker yang diberikan.
    """
    print(f"Mengunduh data untuk {len(tickers)} saham dari {start_date} hingga {end_date} dengan interval {interval}...")
    try:
        # group_by='ticker' akan mengembalikan MultiIndex DataFrame, yang lebih mudah diakses
        data = yf.download(tickers, start=start_date, end=end_date, interval=interval, group_by='ticker')
        if data.empty:
            print("Tidak ada data yang diunduh. Periksa ticker atau rentang tanggal.")
            return pd.DataFrame()
        return data
    except Exception as e:
        print(f"Terjadi kesalahan saat mengunduh data: {e}")
        return pd.DataFrame()

def analyze_market_summary(stock_data):
    """
    Menganalisis data saham untuk menemukan top gainers, top losers, dan top volume.
    """
    if stock_data.empty:
        print("Tidak ada data untuk dianalisis.")
        return

    print("\n--- Analisis Ringkasan Pasar ---")

    # Siapkan DataFrame untuk hasil analisis
    results = []

    # Iterasi melalui setiap ticker dalam MultiIndex DataFrame
    for ticker in stock_data.columns.levels[0]:
        # Cek apakah kolom untuk ticker tersebut ada
        if ticker not in stock_data.columns.levels[0]:
            print(f"Peringatan: Data untuk ticker {ticker} tidak ditemukan.")
            continue
            
        ticker_data = stock_data[ticker]
        
        # Pastikan ada data dan kolom yang diperlukan
        if ticker_data.empty or 'Close' not in ticker_data.columns or 'Volume' not in ticker_data.columns:
            # print(f"Melewatkan {ticker}: Data tidak lengkap.")
            continue

        # Hapus baris dengan NaN di 'Close' atau 'Volume' untuk perhitungan
        ticker_data = ticker_data.dropna(subset=['Close', 'Volume'])

        if ticker_data.empty or len(ticker_data) < 2:
            # print(f"Melewatkan {ticker}: Tidak cukup data setelah dropna.")
            continue

        latest_close = ticker_data['Close'].iloc[-1]
        previous_close = ticker_data['Close'].iloc[-2]
        latest_volume = ticker_data['Volume'].iloc[-1]

        # Hitung perubahan persentase
        change_pct = ((latest_close - previous_close) / previous_close) * 100 if previous_close != 0 else 0

        results.append({
            'Ticker': ticker,
            'Latest Close': latest_close,
            'Previous Close': previous_close,
            'Change (%)': change_pct,
            'Volume': latest_volume
        })
    
    if not results:
        print("Tidak ada data yang valid untuk dianalisis (setelah filter).")
        return

    summary_df = pd.DataFrame(results)
    
    # Urutkan untuk Top Gainers
    top_gainers = summary_df.sort_values(by='Change (%)', ascending=False).head(5)
    print("\nTop 5 Gainers (Berdasarkan Persentase Perubahan Harga):")
    print(top_gainers[['Ticker', 'Latest Close', 'Change (%)']].round(2))

    # Urutkan untuk Top Losers
    top_losers = summary_df.sort_values(by='Change (%)', ascending=True).head(5)
    print("\nTop 5 Losers (Berdasarkan Persentase Perubahan Harga):")
    print(top_losers[['Ticker', 'Latest Close', 'Change (%)']].round(2))

    # Urutkan untuk Top Volume
    top_volume = summary_df.sort_values(by='Volume', ascending=False).head(5)
    print("\nTop 5 Saham Berdasarkan Volume Perdagangan Terbaru:")
    print(top_volume[['Ticker', 'Latest Close', 'Volume']].round(2))


if __name__ == "__main__":
    # --- Konfigurasi Tanggal ---
    end_date = datetime.now() # Tanggal hari ini
    start_date = end_date - timedelta(days=7) # Ambil data 7 hari terakhir
    
    # Format tanggal untuk yfinance
    start_date_str = start_date.strftime('%Y-%m-%d')
    end_date_str = end_date.strftime('%Y-%m-%d')

    # --- Langkah 1: Dapatkan Daftar Ticker Saham ---
    # Penting: Ganti fungsi ini dengan cara Anda mendapatkan daftar ticker yang sebenarnya
    tickers = get_indonesian_stock_tickers()
    
    if not tickers:
        print("Daftar ticker kosong. Silakan sediakan daftar ticker saham.")
    else:
        # --- Langkah 2: Unduh Data Saham ---
        historical_data = get_stock_data(tickers, start_date_str, end_date_str, interval="1d")

        # --- Langkah 3: Analisis dan Tampilkan Ringkasan Pasar ---
        analyze_market_summary(historical_data)
```

### Prediksi Stock Saham

```python
import yfinance as yf
import pandas as pd
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
import numpy as np
import warnings

warnings.filterwarnings("ignore") # Mengabaikan peringatan yang tidak perlu

# --- 1. Ambil Data Stock (BBRI.JK) ---
ticker_symbol = "BBRI.JK"

# Tanggal hari ini
end_date = datetime.now()

# Tanggal mulai (5 tahun ke belakang)
start_date = end_date - timedelta(days=20*365) # Menggunakan 365 hari per tahun untuk perkiraan
                                              # Anda bisa sesuaikan jika perlu lebih presisi dengan tahun kabisat

print(f"Mengunduh data saham {ticker_symbol} dari {start_date.strftime('%Y-%m-%d')} hingga {end_date.strftime('%Y-%m-%d')}...")

try:
    # Mengunduh data harian
    data_bbri = yf.download(ticker_symbol, start=start_date, end=end_date, interval="1d")

    if data_bbri.empty:
        print(f"Tidak ada data yang diunduh untuk {ticker_symbol}. Periksa ticker atau rentang tanggal.")
    else:
        print("\nData BBRI.JK berhasil diunduh. 5 baris pertama:")
        print(data_bbri.head())
        print("\n5 baris terakhir:")
        print(data_bbri.tail())

        # Fokus pada harga penutupan untuk prediksi time series
        ts_data = data_bbri['Close'].dropna()

        if ts_data.empty:
            print("Kolom 'Close' kosong setelah menghapus NaN. Tidak bisa melakukan prediksi.")
        else:
            print(f"\nData Time Series (Harga Penutupan) dari {len(ts_data)} titik data.")

            # --- 2. Prediksi Time Series 1 Bulan ke Depan (Menggunakan ARIMA) ---

            # Tentukan jumlah hari untuk prediksi (sekitar 20 hari kerja dalam sebulan)
            forecast_period_days = 22 # Prediksi untuk sekitar 1 bulan (hari kerja bursa)

            # Membagi data menjadi training dan testing (untuk evaluasi model)
            # Ambil sekitar 80% data untuk training, sisanya untuk testing
            train_size = int(len(ts_data) * 0.8)
            train_data, test_data = ts_data[0:train_size], ts_data[train_size:]

            print(f"\nUkuran data training: {len(train_data)}")
            print(f"Ukuran data testing: {len(test_data)}")

            # --- Membangun dan Melatih Model ARIMA ---
            # Menentukan order (p,d,q) ARIMA. Ini seringkali memerlukan tuning atau analisis ACF/PACF.
            # Untuk demo, kita gunakan order yang umum.
            # p: order AR (jumlah observasi sebelumnya yang digunakan)
            # d: order I (jumlah perbedaan non-musiman yang diperlukan untuk stasioneritas)
            # q: order MA (jumlah kesalahan prediksi sebelumnya yang digunakan)
            
            # Catatan: Pemilihan order ARIMA yang tepat sangat penting dan biasanya melibatkan
            # plot ACF/PACF dan kriteria informasi seperti AIC/BIC.
            # Untuk demo ini, kita gunakan nilai coba-coba.
            order = (5, 1, 0) # Contoh order ARIMA

            print(f"\nMelatih model ARIMA dengan order {order}...")
            try:
                # Inisialisasi dan latih model ARIMA
                # Gunakan data pelatihan saja
                model = ARIMA(train_data, order=order)
                model_fit = model.fit()

                print("\nModel ARIMA berhasil dilatih. Ringkasan model:")
                print(model_fit.summary())

                # --- Membuat Prediksi ---
                # Prediksi pada data testing untuk mengevaluasi model
                start_index_test = len(train_data)
                end_index_test = len(ts_data) - 1
                predictions_test = model_fit.predict(start=start_index_test, end=end_index_test)

                # Menghitung RMSE untuk evaluasi
                rmse = np.sqrt(mean_squared_error(test_data, predictions_test))
                print(f"\nRMSE (Root Mean Squared Error) pada data testing: {rmse:.2f}")

                # Memprediksi untuk 1 bulan ke depan dari akhir seluruh data yang ada
                last_date_in_data = ts_data.index[-1]
                future_dates = pd.date_range(start=last_date_in_data + timedelta(days=1), periods=forecast_period_days, freq='B') # 'B' untuk hari kerja bursa

                # Melatih ulang model dengan seluruh data untuk prediksi masa depan yang lebih baik
                print("\nMelatih ulang model ARIMA dengan seluruh data untuk prediksi masa depan...")
                full_model = ARIMA(ts_data, order=order)
                full_model_fit = full_model.fit()
                
                # Membuat prediksi untuk periode ke depan
                forecast_start_index = len(ts_data)
                forecast_end_index = forecast_start_index + forecast_period_days - 1
                
                forecast = full_model_fit.predict(start=forecast_start_index, end=forecast_end_index)
                
                # Assign indeks tanggal ke hasil prediksi
                forecast.index = future_dates

                print(f"\nPrediksi Harga Penutupan BBRI.JK untuk {forecast_period_days} hari ke depan:")
                print(forecast)

                # --- 3. Visualisasi Hasil Prediksi ---
                plt.figure(figsize=(14, 7))
                plt.plot(ts_data.index, ts_data, label='Harga Aktual BBRI.JK (5 Tahun Terakhir)')
                plt.plot(test_data.index, predictions_test, color='green', linestyle='--', label='Prediksi ARIMA (Data Testing)')
                plt.plot(forecast.index, forecast, color='red', linestyle='--', label=f'Prediksi ARIMA ({forecast_period_days} Hari Ke Depan)')
                
                plt.title(f'Prediksi Harga Penutupan Saham {ticker_symbol} dengan Model ARIMA')
                plt.xlabel('Tanggal')
                plt.ylabel('Harga Penutupan (IDR)')
                plt.legend()
                plt.grid(True)
                plt.show()

            except Exception as arima_e:
                print(f"Terjadi kesalahan saat membangun atau melatih model ARIMA: {arima_e}")

except Exception as e:
    print(f"Terjadi kesalahan umum: {e}")
```

```python
import yfinance as yf
import pandas as pd
from datetime import datetime, timedelta
import numpy as np
import tensorflow as tf
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error # Pastikan ini diimpor
import matplotlib.pyplot as plt

# --- Konfigurasi Global ---
# Ticker saham yang ingin Anda prediksi (misal: BBCA.JK untuk Bank Central Asia)
TICKER_SYMBOL = "BBRI.JK" 

# Jumlah hari historis yang akan digunakan sebagai input untuk prediksi LSTM
N_STEPS = 360 #30 hari x 12 Bulan 

# Rentang waktu data historis yang akan diunduh untuk pelatihan model
HISTORICAL_DAYS = 365 * 5 # 5 tahun

# !!! VARIABEL GLOBAL UNTUK JUMLAH HARI PREDIKSI !!!
NUM_FUTURE_DAYS_TO_PREDICT = 21 

# --- Fungsi Pengambilan Data ---

def get_stock_data(ticker, start_date, end_date, interval="1d"):
    """
    Mengunduh data historis untuk satu ticker saham dari Yahoo Finance.
    """
    print(f"Mengunduh data untuk {ticker} dari {start_date.strftime('%Y-%m-%d')} hingga {end_date.strftime('%Y-%m-%d')}...")
    try:
        data = yf.download(ticker, start=start_date, end=end_date, interval=interval)
        if data.empty:
            print(f"Tidak ada data yang diunduh untuk {ticker}. Periksa ticker atau rentang tanggal.")
            return pd.DataFrame()
        return data
    except Exception as e:
        print(f"Terjadi kesalahan saat mengunduh data untuk {ticker}: {e}")
        return pd.DataFrame()

# --- Fungsi Persiapan Data untuk LSTM ---

def prepare_data_for_lstm(data, n_steps):
    """
    Mempersiapkan data historis untuk input model LSTM.
    Ini melibatkan pemilihan kolom 'Close', normalisasi, dan pembentukan urutan.
    """
    if data.empty or 'Close' not in data.columns:
        print("Data kosong atau kolom 'Close' tidak ada. Tidak dapat menyiapkan data LSTM.")
        return None, None, None, None

    df_close = data[['Close']].copy()
    df_close = df_close.dropna() 

    if len(df_close) < n_steps + 1: 
        print(f"Tidak cukup data setelah pembersihan (hanya {len(df_close)} baris). Dibutuhkan setidaknya {n_steps + 1} baris.")
        return None, None, None, None

    scaler = MinMaxScaler(feature_range=(0, 1))
    scaled_data = scaler.fit_transform(df_close)

    X, y = [], []
    for i in range(n_steps, len(scaled_data)):
        X.append(scaled_data[i-n_steps:i, 0]) 
        y.append(scaled_data[i, 0])          
    
    X, y = np.array(X), np.array(y)

    X = X.reshape(X.shape[0], X.shape[1], 1)

    return X, y, df_close, scaler

# --- Fungsi Pembangunan dan Pelatihan Model LSTM ---

def build_lstm_model(input_shape):
    """
    Membangun arsitektur model LSTM sederhana menggunakan Keras.
    """
    model = tf.keras.Sequential([
        tf.keras.layers.LSTM(units=50, return_sequences=True, input_shape=input_shape),
        tf.keras.layers.Dropout(0.2), 
        tf.keras.layers.LSTM(units=50, return_sequences=False),
        tf.keras.layers.Dropout(0.2),
        tf.keras.layers.Dense(units=1) 
    ])
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model

def train_and_predict(ticker, historical_data, n_steps):
    """
    Melatih model LSTM, melakukan prediksi pada data test, 
    dan memproyeksikan harga untuk beberapa hari ke depan.
    Akan menggunakan variabel global NUM_FUTURE_DAYS_TO_PREDICT.
    """
    X, y, original_df_close, scaler = prepare_data_for_lstm(historical_data, n_steps)

    if X is None or len(X) == 0:
        # Menambahkan RMSE_value ke return
        return None, None, None, None, None, None, None 

    train_size = int(len(X) * 0.8) 
    X_train, X_test = X[:train_size], X[train_size:]
    y_train, y_test = y[:train_size], y[train_size:]
    
    if len(X_train) == 0:
        print(f"Tidak cukup data untuk set pelatihan untuk {ticker}.")
        return None, None, None, None, None, None, None

    print(f"\nMelatih model LSTM untuk {ticker}...")
    model = build_lstm_model(input_shape=(n_steps, 1)) 
    model.fit(X_train, y_train, epochs=50, batch_size=32, verbose=0) 

    # --- Evaluasi Model dan Hitung RMSE ---
    rmse_value = np.nan # Inisialisasi RMSE sebagai NaN
    if len(X_test) > 0:
        test_predictions_scaled = model.predict(X_test, verbose=0) 
        test_predictions_denormalized = scaler.inverse_transform(test_predictions_scaled)
        y_test_denormalized = scaler.inverse_transform(y_test.reshape(-1, 1))
        rmse_value = np.sqrt(mean_squared_error(y_test_denormalized, test_predictions_denormalized))
        print(f"  RMSE (Root Mean Squared Error) pada data test untuk {ticker}: {rmse_value:.2f}")
    else:
        print(f"  Tidak ada data test untuk evaluasi model {ticker}.")
    
    # --- Prediksi N_FUTURE_DAYS_TO_PREDICT Hari ke Depan ---
    future_predictions = []
    future_dates = [] 
    
    current_input_sequence = original_df_close['Close'].values[-n_steps:] 
    
    if np.isnan(current_input_sequence).any():
        print("Peringatan: Data terakhir mengandung NaN. Prediksi masa depan mungkin tidak akurat.")
        return None, None, None, None, None, None, None

    current_input_scaled = scaler.transform(current_input_sequence.reshape(-1, 1))

    last_known_date = original_df_close.index[-1]

    for _ in range(NUM_FUTURE_DAYS_TO_PREDICT): 
        X_predict = current_input_scaled.reshape(1, n_steps, 1)
        predicted_price_scaled = model.predict(X_predict, verbose=0)[0][0]
        future_predictions.append(predicted_price_scaled)
        
        next_date = last_known_date + timedelta(days=1)
        while next_date.weekday() >= 5: 
            next_date += timedelta(days=1)
        future_dates.append(next_date.strftime('%Y-%m-%d')) 
        last_known_date = next_date 

        current_input_scaled = np.append(current_input_scaled[1:], [[predicted_price_scaled]], axis=0)

    future_predictions_denormalized = scaler.inverse_transform(np.array(future_predictions).reshape(-1, 1))

    # Mengembalikan RMSE_value
    return original_df_close, y_test_denormalized, test_predictions_denormalized, future_predictions_denormalized, future_dates, scaler, rmse_value

# --- Fungsi Plot Hasil ---

def plot_results(ticker, original_df, y_test, test_predictions, future_predictions, future_dates, n_steps):
    """
    Memplot harga saham riil, prediksi pada data test, dan prediksi X hari ke depan.
    """
    plt.figure(figsize=(16, 8))
    plt.title(f'Harga Saham {ticker} (Real vs Prediksi LSTM)', fontsize=16)
    plt.xlabel('Tanggal', fontsize=12)
    plt.ylabel('Harga Penutupan (IDR)', fontsize=12)

    train_data_len = len(original_df) - len(y_test)
    train_index = original_df.index[:train_data_len]
    plt.plot(train_index, original_df['Close'][:train_data_len], 
             color='blue', label='Harga Riil (Data Latih)')

    test_actual_index = original_df.index[n_steps:][train_data_len-n_steps:]
    plt.plot(test_actual_index, y_test, color='green', label='Harga Riil (Data Uji)')

    plt.plot(test_actual_index, test_predictions, 
             color='red', linestyle='--', label='Prediksi Model (Data Uji)')

    if len(future_dates) == len(future_predictions):
        future_dates_dt = [datetime.strptime(d, '%Y-%m-%d') for d in future_dates]
        plt.plot(future_dates_dt, future_predictions, 
                 color='purple', linestyle='-', marker='o', 
                 label=f'Prediksi {len(future_dates)} Hari ke Depan') 
    else:
        print("Peringatan: Jumlah tanggal masa depan tidak cocok dengan jumlah prediksi masa depan.")

    plt.legend(fontsize=10) 
    plt.grid(True)
    plt.tight_layout()
    plt.show()


if __name__ == "__main__":
    # --- Konfigurasi Tanggal ---
    end_date = datetime.now() 
    start_date = end_date - timedelta(days=HISTORICAL_DAYS) 
    
    # --- Langkah 1: Unduh Data Saham ---
    historical_data = get_stock_data(TICKER_SYMBOL, start_date, end_date)
    
    if not historical_data.empty:
        # --- Langkah 2: Melatih Model dan Melakukan Prediksi ---
        # Menambahkan variabel rmse_value pada bagian penerimaan hasil
        original_df_close, y_test_denormalized, test_predictions_denormalized, future_predictions_denormalized, future_dates, scaler, rmse_value = \
            train_and_predict(TICKER_SYMBOL, historical_data, N_STEPS)
        
        if original_df_close is not None:
            # --- Langkah 3: Cetak RMSE, Prediksi, dan Plot Hasil ---
            print("\n" + "=" * 50)
            print(f"EVALUASI MODEL UNTUK {TICKER_SYMBOL}")
            print("=" * 50)
            if not np.isnan(rmse_value): # Cek apakah RMSE bukan NaN
                print(f"Root Mean Squared Error (RMSE) pada data uji: {rmse_value:.2f}")
            else:
                print("RMSE tidak dapat dihitung karena tidak ada data uji.")
            print("=" * 50 + "\n")

            print("\n" + "=" * 50)
            print(f"PREDIKSI HARGA {TICKER_SYMBOL} UNTUK {NUM_FUTURE_DAYS_TO_PREDICT} HARI KE DEPAN")
            print("=" * 50)
            for i in range(len(future_dates)):
                print(f"Tanggal: {future_dates[i]} | Harga Prediksi: {future_predictions_denormalized[i][0]:.2f}")
            print("=" * 50 + "\n")

            plot_results(TICKER_SYMBOL, original_df_close, 
                         y_test_denormalized, test_predictions_denormalized, 
                         future_predictions_denormalized, future_dates, N_STEPS)
        else:
            print(f"Tidak dapat membuat plot untuk {TICKER_SYMBOL} karena data tidak memadai atau error dalam prediksi.")
    else:
        print(f"Tidak ada data yang tersedia untuk {TICKER_SYMBOL}. Program dihentikan.")
```

### Dijadikan Function

```python
import yfinance as yf
import pandas as pd
from datetime import datetime, timedelta
import numpy as np
import tensorflow as tf
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

def predict_stock_price(
    ticker_symbol: str,
    historical_data_years: int = 5,
    n_steps: int = 60,
    num_future_days_to_predict: int = 10
):
    """
    Melakukan prediksi harga saham menggunakan model LSTM, berdasarkan input yang diberikan.

    Args:
        ticker_symbol (str): Kode ticker saham (misal: "BBCA.JK").
        historical_data_years (int): Berapa tahun data historis yang akan diunduh untuk training. Default 5 tahun.
        n_steps (int): Berapa lama (jumlah hari) inputan yang dipakai untuk setiap prediksi LSTM. Default 60 hari.
        num_future_days_to_predict (int): Berapa hari ke depan yang ingin diprediksi. Default 10 hari.

    Returns:
        tuple: Mengembalikan tuple berisi:
            - future_predictions_df (pd.DataFrame): DataFrame berisi tanggal dan harga prediksi n hari ke depan.
            - plot_figure (matplotlib.figure.Figure): Objek figure plot chart hasil prediksi.
            - rmse_value (float): Nilai Root Mean Squared Error (RMSE) pada data uji.
            Mengembalikan (None, None, None) jika terjadi error atau data tidak memadai.
    """

    print(f"\n{'='*60}")
    print(f"Memulai Prediksi Saham untuk {ticker_symbol}")
    print(f"Data historis: {historical_data_years} tahun | N_Steps: {n_steps} hari | Prediksi: {num_future_days_to_predict} hari")
    print(f"{'='*60}\n")

    # --- 1. Konfigurasi Tanggal ---
    end_date = datetime.now()
    start_date = end_date - timedelta(days=historical_data_years * 365)

    # --- 2. Mengunduh Data Historis ---
    print(f"Mengunduh data untuk {ticker_symbol} dari {start_date.strftime('%Y-%m-%d')} hingga {end_date.strftime('%Y-%m-%d')}...")
    try:
        data = yf.download(ticker_symbol, start=start_date, end=end_date, interval="1d")
        if data.empty:
            print(f"ERROR: Tidak ada data yang diunduh untuk {ticker_symbol}. Periksa ticker atau rentang tanggal.")
            return None, None, None
    except Exception as e:
        print(f"ERROR: Terjadi kesalahan saat mengunduh data untuk {ticker_symbol}: {e}")
        return None, None, None

    # --- 3. Mempersiapkan Data untuk LSTM ---
    df_close = data[['Close']].copy()
    df_close = df_close.dropna()

    if len(df_close) < n_steps + 1:
        print(f"ERROR: Tidak cukup data setelah pembersihan (hanya {len(df_close)} baris). Dibutuhkan setidaknya {n_steps + 1} baris untuk training LSTM.")
        return None, None, None

    scaler = MinMaxScaler(feature_range=(0, 1))
    scaled_data = scaler.fit_transform(df_close)

    X, y = [], []
    for i in range(n_steps, len(scaled_data)):
        X.append(scaled_data[i-n_steps:i, 0])
        y.append(scaled_data[i, 0])
    X, y = np.array(X), np.array(y)
    X = X.reshape(X.shape[0], X.shape[1], 1)

    # --- 4. Membangun dan Melatih Model LSTM ---
    train_size = int(len(X) * 0.8)
    X_train, X_test = X[:train_size], X[train_size:]
    y_train, y_test = y[:train_size], y[train_size:]

    if len(X_train) == 0:
        print(f"ERROR: Tidak cukup data untuk set pelatihan untuk {ticker_symbol}.")
        return None, None, None

    print(f"Melatih model LSTM untuk {ticker_symbol}...")
    model = tf.keras.Sequential([
        tf.keras.layers.LSTM(units=50, return_sequences=True, input_shape=(n_steps, 1)),
        tf.keras.layers.Dropout(0.2),
        tf.keras.layers.LSTM(units=50, return_sequences=False),
        tf.keras.layers.Dropout(0.2),
        tf.keras.layers.Dense(units=1)
    ])
    model.compile(optimizer='adam', loss='mean_squared_error')
    model.fit(X_train, y_train, epochs=50, batch_size=32, verbose=0)
    print("Model LSTM selesai dilatih.")

    # --- 5. Evaluasi Model dan Hitung RMSE ---
    rmse_value = np.nan
    test_predictions_denormalized = np.array([]) # Inisialisasi kosong
    y_test_denormalized = np.array([]) # Inisialisasi kosong

    if len(X_test) > 0:
        test_predictions_scaled = model.predict(X_test, verbose=0)
        test_predictions_denormalized = scaler.inverse_transform(test_predictions_scaled)
        y_test_denormalized = scaler.inverse_transform(y_test.reshape(-1, 1))
        rmse_value = np.sqrt(mean_squared_error(y_test_denormalized, test_predictions_denormalized))
        print(f"  RMSE (Root Mean Squared Error) pada data test: {rmse_value:.2f}")
    else:
        print("  Tidak ada data test untuk evaluasi RMSE.")

    # --- 6. Prediksi Harga untuk n Hari ke Depan ---
    future_predictions_scaled = []
    future_dates = []

    current_input_sequence = df_close['Close'].values[-n_steps:]
    if np.isnan(current_input_sequence).any():
        print("ERROR: Data terakhir mengandung NaN. Tidak dapat memprediksi masa depan.")
        return None, None, None

    current_input_scaled = scaler.transform(current_input_sequence.reshape(-1, 1))
    last_known_date = df_close.index[-1]

    for _ in range(num_future_days_to_predict):
        X_predict = current_input_scaled.reshape(1, n_steps, 1)
        predicted_price_scaled = model.predict(X_predict, verbose=0)[0][0]
        future_predictions_scaled.append(predicted_price_scaled)

        next_date = last_known_date + timedelta(days=1)
        while next_date.weekday() >= 5: # Lewati Sabtu (5) dan Minggu (6)
            next_date += timedelta(days=1)
        future_dates.append(next_date.strftime('%Y-%m-%d'))
        last_known_date = next_date

        current_input_scaled = np.append(current_input_scaled[1:], [[predicted_price_scaled]], axis=0)

    future_predictions_denormalized = scaler.inverse_transform(np.array(future_predictions_scaled).reshape(-1, 1))

    # Membuat DataFrame hasil prediksi n hari ke depan
    future_predictions_df = pd.DataFrame({
        'Tanggal': future_dates,
        'Harga Prediksi': future_predictions_denormalized.flatten()
    })

    # --- 7. Membuat Plot Chart ---
    fig = plt.figure(figsize=(16, 8))
    plt.title(f'Harga Saham {ticker_symbol} (Real vs Prediksi LSTM)', fontsize=16)
    plt.xlabel('Tanggal', fontsize=12)
    plt.ylabel('Harga Penutupan (IDR)', fontsize=12)

    # Plot data latih
    train_data_len = len(df_close) - len(y_test_denormalized)
    train_index = df_close.index[:train_data_len]
    plt.plot(train_index, df_close['Close'][:train_data_len],
             color='blue', label='Harga Riil (Data Latih)')

    # Plot data uji riil
    test_actual_index = df_close.index[n_steps:][train_data_len-n_steps:]
    plt.plot(test_actual_index, y_test_denormalized,
             color='green', label='Harga Riil (Data Uji)')

    # Plot prediksi pada data uji
    plt.plot(test_actual_index, test_predictions_denormalized,
             color='red', linestyle='--', label='Prediksi Model (Data Uji)')

    # Plot prediksi masa depan
    if len(future_predictions_df) > 0:
        future_dates_dt = [datetime.strptime(d, '%Y-%m-%d') for d in future_predictions_df['Tanggal']]
        plt.plot(future_dates_dt, future_predictions_df['Harga Prediksi'],
                 color='purple', linestyle='-', marker='o',
                 label=f'Prediksi {num_future_days_to_predict} Hari ke Depan')

    plt.legend(fontsize=10)
    plt.grid(True)
    plt.tight_layout()

    print(f"\nPrediksi Saham untuk {ticker_symbol} Selesai.\n")

    return future_predictions_df, fig, rmse_value

if __name__ == "__main__":
    # --- Contoh Penggunaan Fungsi predict_stock_price ---
    # Anda bisa memanggil fungsi ini dengan parameter yang berbeda sesuai kebutuhan Anda.

    # Contoh 1: Prediksi BBCA.JK, data 5 tahun, input 60 hari, prediksi 10 hari
    predictions_df, chart_figure, rmse = predict_stock_price(
        ticker_symbol="ASII.JK",
        historical_data_years=5,
        n_steps=180,
        num_future_days_to_predict=30
    )

    if predictions_df is not None:
        print("\n--- Hasil Prediksi Harga Saham Masa Depan ---")
        print(predictions_df.to_string(index=False)) # Cetak DataFrame tanpa indeks
        print(f"\nRMSE pada data uji: {rmse:.2f}" if not np.isnan(rmse) else "\nRMSE: Tidak dapat dihitung")
        plt.show() # Tampilkan plot
    else:
        print("Gagal mendapatkan prediksi")
```
